```python
from datetime import date
dataframe['Cars_Age'] = date.today().year-dataframe['Year']
dataframe.head()
```

```python
dataframe['Brand']=dataframe.Name.str.split().str.get(0)
dataframe['Model']=dataframe["Name"].str.split().str.get(1)+" "+ dataframe["Name"].str.split().str.get(2)
dataframe.head()
```

```python
dataframe.Kilometers_Driven.head()
```

```python
for col in Num_cols:
    print(col)
    print('Skew: ',round(dataframe[col].skew(),2))
    
    plt.figure(figsize=(15,4))
    plt.subplot(1,2,1)
    plt.ticklabel_format(style='plain')
    dataframe[col].hist(grid=False)
    plt.ylabel("Count")
    plt.subplot(1,2,2)
    sns.boxplot(x=dataframe[col])
    
    plt.show()
```
